```python
##Exploración de datos
import pandas as pd
df = pd.read_csv("PixarMovies.csv")  ##Se importa el dataset

print(df.head()) ##Se muestran las primeras filas
```

```python
df.info()
```

```python
df.describe()
```

```python
###Limpieza
print(df.isnull().sum()) ###Ver cuántos nulos hay

print("\nNo se encontraron valores nulos")
```

```python
###Filtros aplicados
filtro_1 = df[df['RT Score']>= 90]    ###Películas con alto puntaje en Rotten Tomatoes
filtro_2 = df[df['IMDB Score'] >= 8.0]  ###Películas con alta puntuación en IMDB
filtro_3 = df[df['Metacritic Score'] >= 80]  ###Películas con alto puntaje en Metacritic

print("Películas con alto puntaje en Rotten Tomatoes:")
print(filtro_1 [["Movie", "RT Score"]])
print("\nPelículas con alta puntuación en IMDB:")
print(filtro_2 [["Movie", "IMDB Score"]])
print("\nPelículas con alto puntaje en Metacritic")
print(filtro_3 [["Movie", "Metacritic Score"]])
```

```python
##Columnas nuevas
df["Rentabilidad"] = df['Worldwide Gross'] / df['Production Budget']
df["Crítica Promedio"] = (df['RT Score'] + df['IMDB Score'] + df['Metacritic Score']) / 3

print(df[["Movie", "Rentabilidad", "Crítica Promedio"]])
```

```python
promedios_anuales = df.groupby("Year Released")[["Worldwide Gross", "RT Score"]].mean()
print(promedios_anuales)
```

```python
###Insights
Las películas con mejor crítica suelen tener mayor recaudación internacional.
Pixar mantuvo puntuaciones muy altas en los primeros años (2000–2010).
Las películas con más de 8.0 en IMDB suelen ganar más del doble de su presupuesto.
La mayoría de las ganancias vienen del mercado internacional.
Las películas nominadas al Óscar tienen rentabilidad promedio más alta.
```
