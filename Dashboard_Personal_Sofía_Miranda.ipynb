```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("PixarMovies.csv")

##Grafico: Presupuesto vs Ao
plt.figure(figsize=(8,4))
plt.plot(df["Year Released"], df["Production Budget"], color="purple", marker="o")
plt.title("Presupuesto de Produccin por Ao")
plt.xlabel("Ao")
plt.ylabel("Presupuesto (USD)")
plt.grid(True)
plt.savefig("presupuesto_vs_ao.png")
plt.show()

##Grafico: Recaudacin Mundial

plt.figure(figsize=(8,4))
plt.plot(df["Year Released"], df["Production Budget"], color="purple", marker="o")
plt.title("Presupuesto de Produccin por Ao")
plt.xlabel("Ao")
plt.ylabel("Presupuesto (USD)")
plt.grid(True)
plt.savefig("presupuesto_vs_ao.png")
plt.show()

##Grafico: IMDB y RT Score
plt.figure(figsize=(8,4))
plt.plot(df["Year Released"], df["IMDB Score"], label="IMDB", color="red")
plt.plot(df["Year Released"], df["RT Score"], label="Rotten Tomatoes", color="blue")
plt.title("Comparacin de Puntuaciones IMDB vs RT")
plt.xlabel("Ao")
plt.ylabel("Puntuacin")
plt.legend()
plt.grid(True)
plt.savefig("imdb_vs_rt.png")
plt.show()

##Grfico: Ingresos Domsticos vs Internacionales
plt.figure(figsize=(8,4))
plt.plot(df["Year Released"], df["Domestic Gross"], label="Domstico", color="orange")
plt.plot(df["Year Released"], df["International Gross"], label="Internacional", color="teal")
plt.title("Ingresos Domsticos vs Internacionales")
plt.xlabel("Ao")
plt.ylabel("USD")
plt.legend()
plt.grid(True)
plt.savefig("ingresos_dom_vs_inter.png")
plt.show()

##Grfico: Relacin entre Presupuesto y Ganancia
plt.figure(figsize=(6,6))
plt.scatter(df["Production Budget"], df["Worldwide Gross"], color="magenta", alpha=0.7)
plt.title("Relacin Presupuesto - Ganancia Mundial")
plt.xlabel("Presupuesto (USD)")
plt.ylabel("Ganancia (USD)")
plt.grid(True)
plt.savefig("presupuesto_vs_ganancia.png")
plt.show()
```
