```python
import pandas as pd
df = pd.read_csv("PixarMovies.csv")

print("Primeras filas:")
print(df.head())

print("\nÚltimas filas:")
print(df.tail())

print("\nTamaño del dataset (filas, columnas):", df.shape)

print("\nInformación general:")
print(df.info())

print("\nEstadísticas numéricas:")
print(df.describe())
```

```python
###Estadísticas
año_minimo = df['Year Released'].min()
año_maximo = df['Year Released'].max()

promedio_imdb = df['IMDB Score'].mean()

promedio_duracion = df['Length'].mean()

print("\nEstadísticas:")
print(f"Película más antigua: {año_minimo}")
print(f"Película más reciente: {año_maximo}")
print(f"Promedio de IMDB {promedio_imdb:.2f}")
print(f"Promedio duración: {promedio_duracion:.1f}")
```

```python
###Nueva Columna
df["Years since release"] = 2025 -  df['Year Released']

##Muestra solo de la nueva columna

print(df[['Movie','Years since release']])
```

```python
##Hallazgos
- El promedio de duración de las películas de Pixar es de aproximadamente 100 minutos
- El puntaje promedio de IMDB ronda los 8 puntos
- La película más antigua es Toy Story, estrenada en 1995
- La película más reciente es Inside Out o Intensamente, estrenada hace 10 años
```
