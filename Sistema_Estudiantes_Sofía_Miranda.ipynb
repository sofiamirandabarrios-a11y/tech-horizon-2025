```python
#Sistema de estudiantes mejorado
estudiantes = [
    {"nombre": "Ana", "edad": 16, "grado": "11°", "materias": ["Matemáticas", "Arte"], "promedio": 4.5},
    {"nombre": "Luis", "edad": 17, "grado": "12°", "materias": ["Física", "Historia"], "promedio": 3.8},
    {"nombre": "María", "edad": 16, "grado": "11°", "materias": ["Inglés", "Biología"], "promedio": 4.2},
    {"nombre": "Carlos", "edad": 18, "grado": "12°", "materias": ["Matemáticas", "Química"], "promedio": 3.9},
    {"nombre": "Lucía", "edad": 17, "grado": "11°", "materias": ["Arte", "Música"], "promedio": 4.7}
]

def agregar_estudiante():
  nombre = input("Ingrese el nombre del estudiante: ")
  edad = int(input("Ingrese la edad del estudiante: "))
  grado = input("Ingrese el grado del estudiante: ")
  materias = input("Ingrese las materias del estudiante (separadas por comas): ").split(",")
  promedio = float(input("Ingrese el promedio del estudiante: "))

  estudiante = {
      "nombre": nombre,
      "edad": edad,
      "grado": grado,
      "materias": [m.strip() for m in materias],
      "promedio": promedio
  }
  estudiantes.append(estudiante)
  print(f"Estudiante {nombre}, agregado correctamente.\n")

def mostrar_estudiantes():
  print("\n===Lista de Estudiantes===")
  for estudiante in estudiantes:
    print(f"{estudiante['nombre']} ({estudiante['edad']} años, {estudiante['grado']})")
    print(f"  Materias favoritas: {', '.join(estudiante['materias'])}")
    print(f"  Promedio: {estudiante['promedio']}\n")

def buscar_estudiante(nombre):
  for estudiante in estudiantes:
    if estudiante["nombre"].lower() == nombre.lower():
      return estudiante
  return None

def eliminar_estudiante():
  nombre = input("Ingrese el nombre del estudiante a eliminar: ")
  estudiante = buscar_estudiante(nombre)
  if estudiante:
    estudiantes.remove(estudiante)
    print(f"Estudiante {nombre} eliminado correctamente.\n")
  else:
    print("Estudiante no encontrado.\n")

def actualizar_estudiante():
  nombre = input("Ingrese el nombre del estudiante a actualizar: ")
  estudiante = buscar_estudiante(nombre)
  if not estudiante:
    print("Estudiante no encontrado.\n")
    return

  print("Ingrese los nuevos datos del estudiante:")
  nuevo_nombre = input("Nuevo nombre (dejar en blanco para mantener el actual): ")
  nueva_edad = input("Nueva edad (dejar en blanco para mantener el actual): ")
  nuevo_grado = input("Nuevo grado (dejar en blanco para mantener el actual): ")
  nuevas_materias = input("Nuevas materias (separadas por comas, dejar en blanco para mantener las actuales): ")
  nuevo_promedio = input("Nuevo promedio (dejar en blanco para mantener el actual): ")

  estudiante["nombre"] = nuevo_nombre if nuevo_nombre else estudiante["nombre"]
  estudiante["edad"] = int(nueva_edad) if nueva_edad else estudiante["edad"]
  estudiante["grado"] = nuevo_grado if nuevo_grado else estudiante["grado"]
  estudiante["materias"] = [m.strip() for m in nuevas_materias.split(",")] if nuevas_materias else estudiante["materias"]
  estudiante["promedio"] = float(nuevo_promedio) if nuevo_promedio else estudiante["promedio"]
  print("Estudiante actualizado correctamente.\n")

def calcular_promedio_general():
    if not estudiantes:
      return 0
    total = sum(estudiante["promedio"] for estudiante in estudiantes)
    return total / len(estudiantes)

def menu():
  while True:
    print("\n===Menú Principal===")
    print("1. Mostrar Estudiantes")
    print("2. Agregar Estudiante")
    print("3. Buscar Estudiante")
    print("4. Eliminar Estudiante")
    print("5. Actualizar Estudiante")
    print("6. Calcular promedio general")
    print("7. Salir")
    opcion = input("Seleccione una opción: ")

    if opcion == "1":
      mostrar_estudiantes()
    elif opcion == "2":
      agregar_estudiante()
    elif opcion == "3":
      nombre = input("Ingrese el nombre del estudiante a buscar: ")
      estudiante = buscar_estudiante(nombre)
      if estudiante:
        print(f"Estudiante encontrado:\n{estudiante}")
      else:
        print("Estudiante no encontrado.\n")
    elif opcion == "4":
      eliminar_estudiante()
    elif opcion == "5":
      actualizar_estudiante()
    elif opcion == "6":
      print("Promedio general:", calcular_promedio_general())
    elif opcion == "7":
      print("Saliendo del sistema...")
      break
    else:
      print("Opción inválida. Por favor, seleccione una opción válida.\n")

menu()

##
```
